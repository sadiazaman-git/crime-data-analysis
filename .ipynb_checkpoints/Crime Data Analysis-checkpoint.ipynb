**Remove unnecessary columns**

We can exclude the "MINUTE" column since we don't require minute-level granularity. While location-related columns (X, Y, Neighbourhood, Hundred block) won't be utilized in this particular analysis

```python
df.drop(['MINUTE'], axis = 1, inplace=True)
```

```python
df.describe()
```

#### Check for Missing Values

```python
df.info()
```

We can see that we have 530,652 entries, but some columns (HOUR, HUNDRED_BLOCK and NEIGHBOURHOOD) have less, which means that there are missing values.
we have enough data to analyse, so let's remove the null values. 

```python
#drop null values and reset index to sort the index
df.dropna(inplace = True)
df.reset_index(drop = True, inplace = True)
```

```python
#test if there is any null values there.
df.isna().sum()
```

#### DATE

```python
#let's create a new column date to create more features like days of week? 
df['DATE'] = pd.to_datetime({'year':df['YEAR'], 'month':df['MONTH'], 'day':df['DAY']})
```

```python
#(Monday=0 to Sunday=6)
df['DAY_OF_WEEK'] = df['DATE'].dt.dayofweek
```

```python
# Change the index to the colum 'DATE'
df.index = pd.DatetimeIndex(df['DATE'])
```

#### Category

```python
df['TYPE'].value_counts()
```

From the types above, I'm creating more generic categories: Break and Enter, Theft, Vehicle Collision, Othersa

```python
# Create a function to categorize types, using an 'if' statement.
def category(crime_type):
    if 'Theft' in crime_type:
        return 'Theft'
    elif 'Break' in crime_type:
        return 'Break and Enter'
    elif 'Collision' in crime_type:
        return 'Vehicle Collision'
    else:
        return 'Others'
```

```python
# Apply the function and add it as CATEGORY column
df['CATEGORY'] = df['TYPE'].apply(category)
```

### Outlier

```python
# Using idxmax() to find out the index of the max value
df.resample('D').size().idxmax()
```

So the day was 2011-06-15.

```python
# Create a Upper Control Limit (UCL) and a Lower Control Limit (LCL) without the outlier
crimes_daily = pd.DataFrame(df[df['DATE'] != '2011-06-15'].resample('D').size())
crimes_daily['MEAN'] = df[df['DATE'] != '2011-06-15'].resample('D').size().mean()
crimes_daily['STD'] = df[df['DATE'] != '2011-06-15'].resample('D').size().std()
UCL = crimes_daily['MEAN'] + 3 * crimes_daily['STD']
LCL = crimes_daily['MEAN'] - 3 * crimes_daily['STD']

# Plot Total crimes per day, UCL, LCL, Moving-average
plt.figure(figsize=(15,6))
df.resample('D').size().plot(label='Crimes per day')
UCL.plot(color='red', ls='--', linewidth=1.5, label='UCL')
LCL.plot(color='green', ls='--', linewidth=1.5, label='LCL')
crimes_daily['MEAN'].plot(color='yellow', linewidth=2, label='Average')
plt.title('Total crimes per day', fontsize=16)
plt.xlabel('Day')
plt.ylabel('Number of crimes')
plt.tick_params(labelsize=14)
plt.legend(prop={'size':16});
```

We can see some days over the Control Limits, indicating signals. Also, the period of 2003 to 2008 is above the average. Maybe a different Control Limit could be done for that period, but that's ok for now.

### Google Trend Vs Crime

```python
# Importing the data
googletrend = pd.read_csv('data/googletrend.csv', index_col='Month')
googletrend
```

The search index column shows the popularity of the search. A value of 100 is the peak of popularity. Other values are relative to the peak.

```python
# The crime data starts from 2003, but our Google data starts from 2004 and ends in 2017-06. 
# Let's remove 2003 from our crime data and 2017-07.
crimes_month = df[(df['DATE'] > '2003-12-31') & (df['DATE'] < '2017-07-01') ]

# We need to group it by month to compare it to the Google trend.
crimes_month = pd.DataFrame(crimes_month.resample('M').size()) 

# Just renaming the column...
crimes_month.columns = ['Total']

# Dividing the total number of crimes by the maximum value and round them
crimes_month['Crime Index'] = (crimes_month['Total']/crimes_month['Total']
                               .max()*100).astype(int)

crimes_month
```

```python
#Now let's join the two data frames.
crimes_month.index = crimes_month.index.astype(str)
googletrend.index = googletrend.index.astype(str)
crime_trend = pd.concat([crimes_month['Crime Index'],googletrend], axis =1)
crime_trend.head()
```

```python
crime_trend.plot(figsize=(12,6), linewidth=3)
plt.title('Crime Index and Google Trends', fontsize=16)
plt.legend(prop={'size':14})
plt.show()
```

This is interesting. Note that after 2010 there is a lag between crime index and search index. When crime increases, it takes a while until searches increase. When crime index is reaching its local peak, the search start increasing.


## Machine learning model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
prepared_df = df.groupby(['YEAR', 'MONTH','DAY','CATEGORY']).size().rename('crimes').reset_index()
```

```python
prepared_df
```

```python
# Feature engineering and preprocessing
features = prepared_df[['YEAR', 'MONTH', 'DAY', 'CATEGORY']]
labels = prepared_df['crimes']

# Convert categorical features to numerical using one-hot encoding
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), ['CATEGORY'])
    ],
    remainder='passthrough'
)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

# Linear Regression Model
linear_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])

linear_model.fit(X_train, y_train)
linear_predictions = linear_model.predict(X_test)

# Ridge Regression Model
ridge_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', Ridge())
])

ridge_model.fit(X_train, y_train)
ridge_predictions = ridge_model.predict(X_test)

# Lasso Regression Model
lasso_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', Lasso())
])

lasso_model.fit(X_train, y_train)
lasso_predictions = lasso_model.predict(X_test)

# Evaluate Models
def evaluate_model(model_name, predictions, y_true):
    mse = mean_squared_error(y_true, predictions)
    r2 = r2_score(y_true, predictions)
    print(f'{model_name} - Mean Squared Error: {mse}, R-squared: {r2}')

evaluate_model('Linear Regression', linear_predictions, y_test)
evaluate_model('Ridge Regression', ridge_predictions, y_test)
evaluate_model('Lasso Regression', lasso_predictions, y_test)
```

Linear Regression and Ridge Regression show similar performance, with Lasso Regression slightly lagging in terms of MSE and R-squared. Ridge Regression, which includes a regularization term, appears to perform consistently well.

### Cross Validation 

```python
from sklearn.model_selection import cross_val_score, KFold

# Set up K-Fold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Perform cross-validation for Linear Regression
linear_cv_scores = cross_val_score(linear_model, features, labels, scoring='neg_mean_squared_error', cv=kf)
linear_cv_r2_scores = cross_val_score(linear_model, features, labels, scoring='r2', cv=kf)

# Perform cross-validation for Ridge Regression
ridge_cv_scores = cross_val_score(ridge_model, features, labels, scoring='neg_mean_squared_error', cv=kf)
ridge_cv_r2_scores = cross_val_score(ridge_model, features, labels, scoring='r2', cv=kf)

# Perform cross-validation for Lasso Regression
lasso_cv_scores = cross_val_score(lasso_model, features, labels, scoring='neg_mean_squared_error', cv=kf)
lasso_cv_r2_scores = cross_val_score(lasso_model, features, labels, scoring='r2', cv=kf)

# Display the cross-validation results
def display_cv_results(model_name, mse_scores, r2_scores):
    print(f'{model_name} - Mean MSE: {mse_scores.mean()}, Mean R-squared: {r2_scores.mean()}')

display_cv_results('Linear Regression', -linear_cv_scores, linear_cv_r2_scores)
display_cv_results('Ridge Regression', -ridge_cv_scores, ridge_cv_r2_scores)
display_cv_results('Lasso Regression', -lasso_cv_scores, lasso_cv_r2_scores)
```

* The mean MSE and R-squared scores after cross-validation are slightly better for all models, indicating improved generalization to unseen data.
* The consistency between the results before and after cross-validation suggests that the initial assessment was indicative of the models' overall performance.

The cross-validated results shows the models' effectiveness, with Linear and Ridge Regression demonstrating particularly good performance in predicting the number of crimes.

```python
from sklearn.model_selection import learning_curve
from sklearn.model_selection import validation_curve

def plot_learning_curve(model, features, labels, train_sizes=np.linspace(0.1, 1.0, 10), cv=5):
    train_sizes, train_scores, test_scores = learning_curve(
        model, features, labels, train_sizes=train_sizes, cv=cv, scoring='neg_mean_squared_error'
    )

    train_scores_mean = -np.mean(train_scores, axis=1)
    test_scores_mean = -np.mean(test_scores, axis=1)

    plt.figure(figsize=(10, 6))
    plt.plot(train_sizes, train_scores_mean, label='Training Error')
    plt.plot(train_sizes, test_scores_mean, label='Validation Error')

    plt.title('Learning Curve')
    plt.xlabel('Training Examples')
    plt.ylabel('Mean Squared Error')
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()

# Assuming 'random_forest_model' is the trained Random Forest model
plot_learning_curve(ridge_model, features, labels)
```

While the models show good performance, there may still be opportunities for improvement. Consider exploring hyperparameter tuning, feature engineering, or trying more advanced regression models to further enhance predictive accuracy.

```python
from sklearn.model_selection import KFold, GridSearchCV

# Define hyperparameter grid
param_grid = {'regressor__alpha': [0.1, 1, 10, 100]}

# Set up K-Fold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# GridSearchCV for Ridge Regression
ridge_grid = GridSearchCV(ridge_model, param_grid, scoring='neg_mean_squared_error', cv=kf)
ridge_grid.fit(features, labels)

# Get the best hyperparameters
best_alpha = ridge_grid.best_params_['regressor__alpha']

# Update Ridge Regression model with the best hyperparameters
ridge_model.set_params(regressor__alpha=best_alpha)

# Train the refined model on the entire dataset
ridge_model.fit(features, labels)

# Make predictions on the test set
ridge_predictions = ridge_model.predict(X_test)

# Evaluate the refined model
mse = mean_squared_error(y_test, ridge_predictions)
r2 = r2_score(y_test, ridge_predictions)

print(f'Ridge Regression - Mean Squared Error: {mse}, R-squared: {r2}')
```

it suggests that the hyperparameter tuning with Ridge Regression might not be significantly impacting the model's performance on your specific dataset.

Let's consider trying more complex models such as Random Forest Regressor. This model might capture non-linear relationships in the data.

```python
from sklearn.ensemble import RandomForestRegressor

# Random Forest Regression Model
random_forest_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor(random_state=42))
])

# Train the Random Forest model on the entire dataset
random_forest_model.fit(features, labels)

# Make predictions on the test set
random_forest_predictions = random_forest_model.predict(X_test)

# Evaluate the Random Forest model
rf_mse = mean_squared_error(y_test, random_forest_predictions)
rf_r2 = r2_score(y_test, random_forest_predictions)

print(f'Random Forest Regression - Mean Squared Error: {rf_mse}, R-squared: {rf_r2}')
```

* As per the calculation above the RMSE is 0.98 which is rather low as compared to other models and 
* These metrics suggest that the Random Forest Regression model has significantly improved predictive accuracy as compared to the linear regression models. 
* The low MSE indicates that the model is making accurate predictions, and the high R-squared value implies that the majority of the variability in the number of crimes is explained by the model.
* Random Forest Regression model is a strong candidate for predicting the number of crimes in our dataset
